```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rc("font", family='Microsoft YaHei')
```

```python
df=pd.read_excel('C:/Users/nbbzyh/Desktop/新数据ei/互联网指数.xlsx')
df
```

```python
use=df[df.columns[2:12]]
use
```

```python
pca=PCA(n_components=2)
pca.fit(use)
```

```python
pca.components_
```

```python
plt.plot(pca.explained_variance_ratio_,marker='o')
plt.savefig('2.png',dpi=450)
```

```python

pca.explained_variance_ratio_
```

```python
k1_spss = pca.components_.T 
k1_spss
```

```python
weight = (np.dot(k1_spss, pca.explained_variance_ratio_)) / np.sum(pca.explained_variance_ratio_)
print('weight:',weight)
```
